## < Feature Selection Metric 7 > 
### `F1-Measure`    
부연) 특정한 단어를 갖고 comfusion matrix를 만들었을 때, 그 comfusion matrix를 기준으로 해서 Recall과 Precision을 계산할 수 있습니다.
- Expected accuracy of a simple classifier built from the single feature   

$$ F1(w) = \frac{2 \times Recall(w) \times Precision(w)}{Recall(w) \times Precision(w)} $$  

$$ Recall(w) = \frac{N(Pos, w)}{N(Pos, w) + N(Pos, \bar{w}} , Precision(w) = \frac{N(Pos, w)}{N(Pos, w) + N(Neg, w)}  $$   

Recall은 실제로 Positive 중에서 얼마나 $w$가 등장했느냐에 대한 것이고, Precision은 $w$가 등장한 문서들 중에서 Positive인 것으로 정리할 수 있겠습니다.

- By doing some arithmetic operations, we can derive  

$$ F1(w) = \frac{2 \times N(Pos, w)}{N(Pos) + N(w)} $$  

- In F1 measure, negative features are **devalued** compared to positive features

위에서 각각의 F1-Measure를 보시면 특정한 범주를 기준으로 했을 때의 값은 그렇게까지 좋은 평가 지표가 아니라는 것을 확인할 수 있습니다.
