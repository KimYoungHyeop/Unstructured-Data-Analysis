### Homomorphism

스크린샷 2022-06-04 오후 1.13.14.png

- Want to preserve $\frac{P(k|ice)}{P(k|steam)}$ using $F((w_i-w_j)^T\tilde{w}_k)$ = $\frac{P_{ik}}{P_{jk}}$   

 $\frac{P(solid|ice)}{P(solid|steam)} = F((ice-steam)^Tsolid)$    

 $\frac{P(solid|steam)}{P(solid|ice)} = F((steam-ice)^Tsolid)$

$$F((ice-steam)^Tsolid) = \frac{P(solid|ice)}{P(solid|steam)} = \frac{1}{F((steam-ice)^Tsolid)}     $$  

<center>부연) ice와 steam의 위치를 바꿨을 때, 위치를 바꾸고 난 함수는 원래 함수의 역수가 되어야 한다는 의미입니다.</center>

> $$(ice-steam)^Tsolid = -(steam-ice)^Tsolid$$   
> <center>inverse element of addition(덧셈에 대한 항등원!)</center>   

<center>Input이 되는 겁니다.</center>

> $$F((ice-steam)^Tsolid) = \frac{1}{F((steam-ice)^Tsolid)}     $$  
> <center>inverse element of multiplication(곱셈에 대한 항등원!)</center>  

<center>함수의 Output이 되는 겁니다.</center>

- Homomorphism preserves an operation, which it turn preserves the inverse element 
- Need a homomorphism from $(\mathbb{R}, +)$ to $(\mathbb{R}_{>0}, \times)$  

즉, 함수의 인자에 대한 실수 공간 상에서의 덧셈의 항등원 관계가   
함수의 출력 공간 상에서는 0보다 큰 영역에서의 곱셈에 대한 항등원으로 바뀌어야 한다(매핑)는 뜻입니다.

- Function F: homomorphism that maps $(\mathbb{R}, +)$ to $(\mathbb{R}_{>0}, \times)$  

$$ w_i^T\tilde{w}_k = (w_i - w_j)^T\tilde{w}_k + w_j^T\tilde{w}_k  $$   

$$ F(w_i^T\tilde{w}_k) = F((w_i - w_j)^T\tilde{w}_k + w_j^T\tilde{w}_k) = F((w_i - w_j)^T\tilde{w}_k) \times F(w_j^T\tilde{w}_k) $$  

$$ F((w_i - w_j)^T\tilde{w}_k) = \frac{F(w_i^T\tilde{w}_k)}{F(w_j^T\tilde{w}_k)} = \frac{P_{ik}}{P_{jk}} $$   

- Finally, we can drive that $F(x) = exp(x)$

### Solution
- We know that $F(x) = exp(x)$ and $F((w_i - w_j)^T\tilde{w}_k) = \frac{F(w_i^T\tilde{w}_k)}{F(w_j^T\tilde{w}_k)} = \frac{P_{ik}}{P_{jk}}$

$$exp(w_i^T\tilde{w}_k - w_j^T\tilde{w}_k) = \frac{exp(w_i^T\tilde{w}_k)}{exp(w_j^T\tilde{w}_k)}   $$  

<center>⬇</center> $$ w_i^T\tilde{w}_k = logP_{ik} = logX_{ik} - logX_i $$  

<center>⬇</center> $$ w_i^T\tilde{w}_k = logP_{ik} = logX_{ik} - logX_i $$  
$$ w_i^T\tilde{w}_k = logX_{ik} - b_i - \tilde{b}_k $$  
$$w_i^T\tilde{w}_k + b_i + \tilde{b}_k = logX_{ik}$$

### Objective Function
- A least squared objective function   

$$ J = \sum_{i,j = 1}^V (w_i^T\tilde{w}_j + b_i + \tilde{b}_j - logX_{ij})^2$$  
<center>⇓</center>
$$\sum_{i,j = 1}^V f(X_{ij})(w_i^T\tilde{w}_j + b_i + \tilde{b}_j - logX_{ij})^2$$

where $f$ has the following desiderata:
1. $f(0)$ = $0$ 
2. $f(x)$ should be non-decresing so that rare co-occurrences are not overweighted.
3. $f(x)$ should be relatively small for large values of $x$, so that frequent co-occurrences are not overweighted.  

2번, 3번을 이용해서 수식의 $f(x_{ij})$ 부분을 보면,  
high-frequency words에 대해 학습에 대한 가중치를 내려주는 역할을 수행하게 됨을 알 수 있습니다.


 
$$ J = \sum_{i,j=1}^Vf(X_{ij})(w_i^T\tilde{w}_j + b_i + \tilde{b}_j - logX_{ij})^2 $$
