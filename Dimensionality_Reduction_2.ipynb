# Feature Selection

## < Artificial Data Set >
### 10 Documents with 10 Terms
- Binary classification/categorization problem
- 6 positive documents & 4 negative documents
- Binary Term-Document matrix 

아래 그림을 보시면 D1부터 D6까지가 positive label이 되겠고, D7부터 D10까지가 negative label을 가지고 있는 Document입니다. Matrix에 표현된 숫자는 Binary representation으로써 해당하는 Term이 한 번이라도 등장했으면 1, 아니라면 0으로 표현됩니다.  

여기서의 질문은 "Positive/Negative를 잘 구분하기 위해서 가장 유용한 Term이 무엇인가를 찾아주세요."입니다.   

두 범주를 구분하려면 상식적으로 한 쪽 범주에만 몰아서 등장하는 것이 좋겠는데요, 그런 관점에서 보면 계산을 하지 않더라도 Term 1은 Positive 범주에만 등장했고 Term 2는 Negative 범주에만 등장을 했으니 이 두 Term은 중요할 가능성이 높은 반면에 Term 3은 둘 다 등장(마치 관사 The 같은 느낌!)했으니 중요도가 낮을 가능성이 높습니다.

## < Feature Selection Metric 1-4 >  
이제부터 Feature Selection 방식에서의 filter 방식 관점으로 선택할 수 있는 총 열 가지의 metric에 대한 설명이 나옵니다.   
뒤로 갈수록 computation은 더 많아지는 대신에 변별력은 앞부분 보다 높아진다고 보시면 될 것 같습니다.

### `Document frequency (DF)`
- Simply count the number of total documents in which a word $w$ is presented  

$$ DF(w) = N_D(w) $$

### `Accuracy (Acc)`
- Expected accuracy of a simple classifier built from the single feature   

$$ Acc(w) = N(Pos, w) - N(Neg, w) $$

스크린샷 2022-06-13 오전 9.04.34.png

- .
  - For Term 1: $N(Pos, w)$ = 6, $N(Neg, w)$ = 0, $Acc(w)$ = 6
  - For Term 2: $N(Pos, w)$ = 0, $N(Neg, w)$ = 4, $Acc(w)$ = -4
  - For Term 3: $N(Pos, w)$ = 6, $N(Neg, w)$ = 4, $Acc(w)$ = 2 

### `Accuracy ratio (AccR)`
- Expected accuracy of a simple classifier built from the single feature   

$$ AccR(w) = \left|\frac{N(Pos, w)}{N(Pos)} - \frac{N(Neg, w)}{N(Neg)}\right| $$

스크린샷 2022-06-13 오전 9.04.34.png

- .
  - For Term 1: $\frac{N(Pos, w)}{N(Pos)}$ = $\frac{6}{6}$ = 1, $\frac{N(Neg, w)}{N(Neg)}$ = $\frac{0}{4}$ = 0, $AccR(w)$ = 1
  - For Term 2: $\frac{N(Pos, w)}{N(Pos)}$ = $\frac{0}{6}$ = 0, $\frac{N(Neg, w)}{N(Neg)}$ = $\frac{4}{4}$ = 1, $AccR(w)$ = 1
  - For Term 3: $\frac{N(Pos, w)}{N(Pos)}$ = $\frac{6}{6}$ = 1, $\frac{N(Neg, w)}{N(Neg)}$ = $\frac{4}{4}$ = 1, $AccR(w)$ = 0
