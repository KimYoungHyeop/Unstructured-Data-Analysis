## < N-Grams >

### N-Gram-based Language Models in NLP
- Use the previous N-1 Words in a sequence to predict the next word

$$ P(w_n|w_{n-1}, w_{n-2},…, w_1) = \frac{P(w_n, w_{n-1}, w_{n-2},…, w_1)}{p(w_{n-1}, w_{n-2},..., w_1)}$$  

- Q) One of the hottest topics in artificial intelligence is deep _____
    - blue vs. frying vs. learning ?

### N-Gram in Text Mining
- Some phrases are very useful in text clustering/categorization!
    - Six sigma, supply chain management, big data, etc.  

- Term-frequency for n-grams can be utilzed.
- Domain-dependent  

     부연) 두 단어 이상으로 이루어진 phrase(복합어)가 유용한 의미를 갖는 경우가 있습니다. 
     'big data'처럼 'big'과 'data'로 나뉘면 각각의 본질적인 의미로 받아들여지겠지만, 'big data'로 이루어진 두 phrase는 그 자체로써 하나의 관용어구로 사용되고 있기 때문에 이런 상황에서는 토큰을 단어 하나 단위로 보기 보다는 N-Gram 기준으로 보는 것이 utilze하다는 의미입니다.
