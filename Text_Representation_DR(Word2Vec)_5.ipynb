## < Word2Vec >   

이전에 커밋한 NNLM에 경우 순차적으로 단어들이 주어졌을 때 단어의 다음 단어가 무엇인지를 예측하는 Neural Network를 만들었는데,  
Word2Vec는 주변 단어를 토대로 예측하고자 하는 단어를 예측하거나(Continuous bag-of-words) 또는 하나의 단어를 토대로 주변에 있는 단어를 예측하는 방식(Skip-gram)이 존재합니다. 


### Two Architectures
- Continuous bag-of-words (CBOW) vs. Skip-gram

CBOW는 위 그림처럼 t번째 단어(*w(t)*)를 예측하는 데에 있어서 Neural Network의 구조 자체가 t번째 단어의 앞뒤 두 개씩을 토대로 t번째 단어를 예측하는 방법론입니다.  
Skip-gram은 반대로 t번째 단어를 토대로 Neural Network의 구조 자체가 t-2, t-1, t+1, t+2 번째 단어를 예측하는 방법론입니다.  

Skip-gram은 gradient를 받을 때 주변 단어들의 정보들을 모두 하나의 단어를 업데이트하는 데에 gradient를 받습니다. 따라서 퍼포먼스는 상대적으로 Skip-gram이 더 낫습니다.  

### Learning representations: Skip-gram approach 
Neural Network이기는 하나 가장 기본적인 구조라서 activation function이 없는 것이 포인트입니다.
- Predict surrounding words in a window of length m of every word
  
### Objective function
- Maximize the log probability of any context word given the current center word  

$$ J(θ) = \frac{1}{T}\sum_{t=1}^{T} \sum_{-m≤j≤m,j≠0}log(w_{t+j}|w_t) $$  

- where $θ$ represents all variables we optimize

$$ p(c|w) = \frac{e^{x_w^Tv_c}}{\sum_{k=1}^Ke^{x_w^Tv_k}} $$
