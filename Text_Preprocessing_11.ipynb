## < Language Modeling >
### Bigram model
- Condition on the previous word    
부연) 컨디션은 바로 '이전' 단어에만 영향을 받는다는 것. 밑에 수식은 i 번째 단어가 첫번째부터 i-1 번째 단어를 보고 발생할 확률을 근사 시키면 i-1 번째 단어가 주어졌을 때 i 번째 단어가 등장할 확률로 보면 됩니다. 다만 보시다시피 이걸 통해서 만들어도 그럴듯하게 만들어지지는 않습니다.

$$P(w_i|w_1w_2...w_{i-1}) \approx P(w_i|w_{i-1}) $$

> outside, new, car, parking, lot, of, the, agreement, reached  
> this, would, be, a, record, november

### N-gram models
- can extend to trigrams, 4-grams, 5-grams
    - In sufficient model of language because language has long-distance dependencies
    - "**The computer** when I had just put into the machine room on the fifty floor **crashed**."  

     부연) 이렇게 증가를 시키더라도 langugage 자체가 long-distance dependencies를 가지고 있을 때는 만들어 내기 쉽지 않습니다. 'crashed'의 지칭하는 대상은 'The computer'인데 예를 들면 5-grams를 썼을 때 crashed 앞에 단어 'on', 'the', 'fifth', 'floor'만 주어지고 crashed를 맞추게 하면 앞부분에 context가 없기 때문에 맞추기 어려운 상황이 됨을 짐작할 수 있습니다.

- We can often get away with N-gram models

### Recurrent Neural Network (RNN)-based Language Model
- Character-level RNN (Korean)
