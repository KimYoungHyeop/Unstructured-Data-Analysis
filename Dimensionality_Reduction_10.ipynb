### Cost Function for a Low-dimensional Representation
- Kullback-Leibler divergence
   - A non-symmetric measure of the difference between two probability distribution P and Q    

$$ Cost = \sum_iKL(P_i||Q_i) = \sum_i\sum_jP_{j|i}log\frac{P_{j|i}}{q_{j|i}}  $$   

- Gradient   

$$ \frac{∂C}{∂y_i} = 2\sum_j(y_j-y_i)(P_{j|i}-q_{j|i}+P_{i|j}-q_{i|j})  $$  


- Update the coordinate in the lower dimension to minimize the Cost function  
- Gradient
   - Differencing Cost is tedious because $y_k$ affect $q_{ij}$ via the normalized term in Eq. (3), but the result is simple
   - The gradient has a surprisingly simple form  

- Gradient update with a momentum term   

$$ y^{(t+1)} = y^{(t)} + \eta\frac{∂C}{∂y} + α(t)\left( y^{(t)}-y^{(t-1)} \right)  $$

### From the original paper
