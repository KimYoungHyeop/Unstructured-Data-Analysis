### GloVe
- Based on matrix factorization method
- http://nlp.stanford.edu/projects/glove/
- Notations

> $X ∈ \mathbb{R}^{V × V}$ word co-occurrence matrix  
> $X_{ij}$ frequency of word $i$ co-occurrence with word $j$  
> $X_i = \sum_{k}^V X_{ik}$ total number of occurrences of word $i$ in corpus   
> $P_{ij} = P(j|i) = \frac{X_{ij}}{X_i}$ a.k.a. probability of word $j$ occurring within the context of word $i$   
> $w ∈ \mathbb{R}^{d}$ a word embedding of dimension $d$   
> $\tilde{w} ∈ \mathbb{R}^{d}$ a context word embedding of dimension $d$

### Motivation

스크린샷 2022-06-03 오후 12.55.11.png

- For words $k$ related ice but not steam (solid), the radio $P_{ik}/P_{jk}$ is large  
- For words $k$ related steam but not ice (gas), the radio $P_{ik}/P_{jk}$ is small   
- For words $k$ that are either related to both ice and steam, or to neither, the ratio should be close to l

### Formulation
- Express the relationship among three words using a function F  
  
$$ F(w_i, w_j, \tilde{w}_k) = \frac{P_{ik}}{P_{jk}} $$  

- Relationship between $w_i$ and w_j is expressed by subtraction   

$$ F(w_i - w_j, \tilde{w}_k) = \frac{P_{ik}}{P_{jk}} $$

- Inner product is used to link $\tilde{w}_k$ with $w_i$ and $w_j$   


$$ F((w_i - w_j)^T \tilde{w}_k) = \frac{P_{ik}}{P_{jk}} $$
